```python
from kiteconnect import KiteConnect
import logging
import json

logging.basicConfig(level=logging.DEBUG)

api_key = ""
access_token = ""

with open("../../env.json", "r") as json_file:
    json_content = json.load(json_file)
    api_key = json_content["kite_api"]["api_key"]
    access_token = json_content["kite_api"]["access_token"]
# print(api_key)
# print(access_token)

kite = KiteConnect(api_key=api_key)
kite.set_access_token(access_token)

# print(kite.instruments())

from_date = "2000-01-01"
to_date = "2004-01-01"
instrument_token = "738561"

historical_data = kite.historical_data(instrument_token, from_date, to_date,"day")
print(historical_data)
```
